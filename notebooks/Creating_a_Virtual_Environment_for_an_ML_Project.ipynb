# Creating a Virtual Environment for an ML Project

A practical guide to creating, activating, and managing a virtual environment for a machine learning project.

````python
import sys
from pathlib import Path

print(f'Python version: {sys.version}')
print(f'Executable: {sys.executable}')
print(f'Project root: {Path.cwd()}')
````

## 1. Check Python Version

Before creating a virtual environment, confirm that a compatible Python version is installed.

```bash
python --version
python3 --version
```

If you are working on Windows PowerShell, `python --version` is usually enough. This should show Python 3.9+ for a smooth ML workflow.

## 2. Create a Project Virtual Environment

Create an isolated environment inside the project directory:

```bash
python -m venv venv
```

Example project structure after creation:

```text
ml-project/
├── venv/
├── src/
├── data/
├── models/
├── requirements.txt
└── README.md
```

The `venv/` folder should not be committed to Git.

## 3. Activate and Deactivate the Environment

On macOS/Linux:

```bash
source venv/bin/activate
```

On Windows:

```powershell
venv\Scripts\activate
```

When you are finished working, leave the environment with:

```bash
deactivate
```

After activation, your prompt usually shows `(venv)` to indicate that you are using the isolated environment.

## 4. Install ML Dependencies

Install the libraries your ML project needs inside the active virtual environment:

```bash
pip install pandas numpy scikit-learn matplotlib seaborn joblib
```

You can verify what is installed with:

```bash
pip list
```

These packages live inside the virtual environment and do not affect your global Python installation.

## 5. Capture Exact Package Versions

After installing the dependencies, freeze the environment into a requirements file:

```bash
pip freeze > requirements.txt
```

This records exact versions such as:

```text
pandas==2.1.0
numpy==1.24.3
scikit-learn==1.3.0
matplotlib==3.7.2
seaborn==0.12.2
joblib==1.3.1
```

Pinned versions are important for reproducibility.

## 6. Recreate the Environment from requirements.txt

A new environment can be recreated by installing everything from the pinned requirements file:

```bash
pip install -r requirements.txt
```

This is the fastest way to rebuild the same environment on another machine or after a fresh setup.

## 7. Verify Environment Isolation

Use these checks to confirm that your terminal or notebook is pointing to the isolated environment rather than the global interpreter:

```bash
pip list
which python   # macOS/Linux
where python   # Windows
```

You can also check the active interpreter from Python:

```python
import sys
import shutil

print(sys.executable)
print(shutil.which('python'))
```

If these paths point inside `venv`, your environment isolation is working correctly.
