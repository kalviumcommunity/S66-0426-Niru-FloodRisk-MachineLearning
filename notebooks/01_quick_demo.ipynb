# Quick Demo: Flood Risk Prediction Pipeline

This notebook demonstrates the complete ML pipeline in action:
1. Load and explore data
2. Train a model
3. Evaluate performance
4. Make predictions

## Setup

```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Add project to path
project_root = Path().resolve().parent
sys.path.insert(0, str(project_root))

from src import config
from src.data_preprocessing import preprocess_data
from src.train import train_and_save
from src.evaluate import evaluate_model

print("✓ All imports successful")
```

## Step 1: Load and Explore Data

```python
# Load raw data
raw_data = pd.read_csv(project_root / config.RAW_DATA_PATH)

print(f"Data shape: {raw_data.shape}")
print(f"\nColumn names and types:")
print(raw_data.dtypes)
print(f"\nFirst few rows:")
display(raw_data.head(10))
```

## Step 2: Data Statistics

```python
print("Data Statistics:")
print(raw_data.describe())

print(f"\nTarget distribution:")
print(raw_data[config.TARGET_COLUMN].value_counts())
print(f"\nClass proportions:")
print(raw_data[config.TARGET_COLUMN].value_counts(normalize=True))
```

## Step 3: Preprocess Data

```python
# Run preprocessing pipeline
X_train, X_test, y_train, y_test = preprocess_data(project_root / config.RAW_DATA_PATH)

print(f"\nPreprocessing complete!")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
```

## Step 4: Train Model

```python
# Train and save model
model, feature_importance = train_and_save(X_train, y_train, project_root / config.MODEL_PATH)

print(f"\nTop 5 Important Features:")
print(feature_importance.head())
```

## Step 5: Evaluate Model

```python
# Evaluate model
results = evaluate_model(model, X_train, X_test, y_train, y_test)

print(f"\nModel Performance Summary:")
print(f"Training Accuracy:  {results['train_metrics']['accuracy']:.4f}")
print(f"Testing Accuracy:   {results['test_metrics']['accuracy']:.4f}")
print(f"Testing Precision:  {results['test_metrics']['precision']:.4f}")
print(f"Testing Recall:     {results['test_metrics']['recall']:.4f}")
print(f"Testing F1-Score:   {results['test_metrics']['f1']:.4f}")
```

## Step 6: Visualize Results

```python
# Plot confusion matrix
cm = results['confusion_matrix']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Confusion Matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False)
axes[0].set_title('Confusion Matrix (Test Set)')
axes[0].set_ylabel('True Label')
axes[0].set_xlabel('Predicted Label')
axes[0].set_xticklabels(['No Flood', 'Flood'])
axes[0].set_yticklabels(['No Flood', 'Flood'])

# Feature Importance
top_features = feature_importance.head(10)
axes[1].barh(range(len(top_features)), top_features['importance'])
axes[1].set_yticks(range(len(top_features)))
axes[1].set_yticklabels(top_features['feature'])
axes[1].set_xlabel('Importance')
axes[1].set_title('Top 10 Important Features')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()

print("✓ Visualizations complete!")
```

## Step 7: Make a Prediction

```python
# Make prediction on a sample
sample = X_test.iloc[0:1]  # Take first test sample
prediction = model.predict(sample)[0]
probability = model.predict_proba(sample)[0]

print(f"Sample features:")
print(sample)

print(f"\nPrediction: {'FLOOD' if prediction == 1 else 'NO FLOOD'}")
print(f"No Flood Probability: {probability[0]:.2%}")
print(f"Flood Probability: {probability[1]:.2%}")
```
