# Creating an ML Project Folder Structure

In machine learning, code is only one part of the system. You are not just writing a script that trains a model, you are building a structured pipeline that ingests raw data, transforms it, trains models, evaluates them, saves artifacts, and produces predictions in a reproducible way.

## Why Folder Structure Matters

A well-designed folder structure reflects clear thinking. It separates responsibilities, prevents accidental data corruption, improves collaboration, and makes your project scalable. A messy structure leads to hidden dependencies, duplicated code, silent errors, and pipelines that only work on one person's laptop.

This is not cosmetic organization. This is system design.

## What This Lesson Supports

In this lesson, you will learn how to design and implement a professional-grade ML project folder structure that supports:

- Clear separation of concerns
- Reproducibility
- Artifact management
- Collaboration
- Clean integration between training and prediction pipelines

## The Core Principle: Separation of Concerns

A structured ML repository follows one fundamental rule: each folder has a single responsibility.

Raw data should never live next to trained models. Exploratory notebooks should not contain production logic. Source code should not be mixed with experiment outputs.

When responsibilities are separated, you gain:

- Easier debugging
- Clear data flow
- Safer experimentation
- Predictable artifact management
- Better onboarding for new team members

## The Standard ML Project Structure

A clean and scalable ML project typically follows this structure:

```text
ml-project/
├── data/
│   ├── raw/
│   ├── processed/
│   └── external/
├── notebooks/
├── src/
│   ├── __init__.py
│   ├── config.py
│   ├── data_preprocessing.py
│   ├── feature_engineering.py
│   ├── train.py
│   ├── evaluate.py
│   ├── predict.py
│   └── persistence.py
├── models/
├── reports/
├── logs/
├── requirements.txt
├── README.md
└── main.py
```

Now let’s understand what each folder represents and why it exists.

## The data/ Folder: Protecting Your Source of Truth

The `data/` directory is your foundation. It must be organized carefully.

### data/raw/
This folder contains original, untouched datasets.

- Never modify files in this folder.
- Never overwrite raw data.
- Treat it as immutable ground truth.

If you alter raw data, you lose reproducibility. You can no longer trace how your features were created.

### data/processed/
This folder contains cleaned or feature-engineered datasets.

These files are generated by your preprocessing scripts and can always be recreated from raw data. They are not sacred.

### data/external/
Optional. Used for third-party datasets or reference files not originally collected by your system.

Separation between raw and processed data prevents accidental leakage and enforces discipline in your workflow.

## The src/ Folder: Where Production Logic Lives

The `src/` folder contains your actual ML pipeline code. This is not where you experiment. This is where structured, reusable, importable logic lives.

Each file should have one responsibility.

### config.py
Stores constants and configuration variables such as:

- RANDOM_STATE
- TEST_SIZE
- FILE PATHS
- TARGET COLUMN

Centralizing configuration prevents hardcoded values across files.

Example:

```python
RANDOM_STATE = 42
TEST_SIZE = 0.2
TARGET_COLUMN = "Churn"
DATA_PATH = "data/raw/telco_churn.csv"
```

### data_preprocessing.py
Responsible for:

- Loading data
- Handling missing values
- Splitting train/test

No model logic should live here.

### feature_engineering.py
Responsible for:

- Encoding
- Scaling
- Derived features
- Building sklearn Pipelines

This module must support both training and prediction.

### train.py
Responsible for:

- Model initialization
- Hyperparameter tuning
- Training execution
- Saving trained model artifacts

This file does not evaluate on test data beyond what is necessary for training decisions.

### evaluate.py
Responsible for:

- Computing metrics
- Generating evaluation reports
- Comparing against baseline

Separation from training prevents accidental evaluation on training data.

### predict.py
Responsible for:

- Loading saved artifacts
- Validating new input
- Applying preprocessing pipeline
- Generating predictions

This file must never refit anything.

### persistence.py
Handles:

- Saving models
- Loading models
- Versioning artifacts

Keeping persistence isolated avoids scattered saving logic.

## The notebooks/ Folder: Exploration Only

Jupyter notebooks belong here.

Notebooks are for:

- EDA
- Visualization
- Hypothesis testing
- Experimentation

They are not for production logic.

Once logic stabilizes, it should move into the `src/` folder.

## The models/ Folder: Artifact Management

This folder stores:

- Trained model files (`.pkl`, `.joblib`)
- Saved preprocessing pipelines
- Versioned artifacts

Example:

```text
models/
  random_forest_v1.pkl
  preprocessing_pipeline_v1.pkl
```

Never retrain on every prediction. Models must be saved and loaded.

## The reports/ Folder: Evaluation Outputs

The `reports/` folder contains:

- evaluation_report.json
- confusion_matrix.png
- performance_plots.png

Evaluation artifacts should be stored separately from models and logs.

## The logs/ Folder: Experiment Tracking

Every experiment should log:

- Algorithm
- Hyperparameters
- Metrics
- Timestamp

Example:

```text
logs/
  experiment_log.csv
```

Tracking experiments prevents guesswork and enables reproducibility.

## Root-Level Files

### requirements.txt
Contains pinned library versions:

```text
pandas==2.1.0
scikit-learn==1.3.0
numpy==1.24.3
joblib==1.3.1
```

This ensures consistent environments.

### README.md
Explains:

- Problem statement
- Dataset
- Approach
- Results
- How to run the project

Without a README, your project is incomplete.

### main.py
Optional entry point that runs the full pipeline end-to-end.

## Training vs Prediction Structure

A professional ML project must clearly separate:

### Training Flow
Raw data → preprocessing → feature engineering → model fitting → artifact saving → evaluation

### Prediction Flow
New data → load artifacts → transform → predict

Folder structure should make this separation obvious.

## Common Structural Mistakes

Avoid the following:

- Everything inside one notebook
- No separation between raw and processed data
- Saving models inside src/
- Hardcoded absolute file paths
- No logs folder
- No requirements.txt

If your project only works on your laptop, the structure is wrong.

````python
# Create the structure step-by-step from the project root
from pathlib import Path

project_root = Path('.')
folders = [
    project_root / 'data' / 'raw',
    project_root / 'data' / 'processed',
    project_root / 'data' / 'external',
    project_root / 'notebooks',
    project_root / 'src',
    project_root / 'models',
    project_root / 'reports',
    project_root / 'logs',
]

for folder in folders:
    folder.mkdir(parents=True, exist_ok=True)

for filename in ['README.md', 'requirements.txt']:
    file_path = project_root / filename
    if not file_path.exists():
        file_path.touch()

config_path = project_root / 'src' / 'config.py'
if not config_path.exists():
    config_path.touch()

print('Project structure created successfully.')
````
