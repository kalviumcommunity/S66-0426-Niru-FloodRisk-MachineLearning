## Version Pinning Strategies

There are three common ways to specify versions:

### Exact Version (Strict Pinning)

```text
scikit-learn==1.3.0
```

Guarantees exact reproducibility. Use this for ML sprint submissions.

### Minimum Version

```text
scikit-learn>=1.3.0
```

Allows newer versions but guarantees at least the minimum. Reduces reproducibility.

### Compatible Version

```text
scikit-learn~=1.3.0
```

Allows patch updates but restricts major/minor upgrades. Useful in production.

**For this sprint, use exact version pinning with `==`.** Reproducibility is more important than flexibility.

## How to Generate requirements.txt

### Method 1: Manual Specification (Recommended)

Write the file explicitly with only the libraries you directly use:

```text
pandas==2.1.0
numpy==1.24.3
scikit-learn==1.3.0
joblib==1.3.1
```

This keeps the file clean and reflects intentional design.

### Method 2: Freeze Installed Packages

If you already installed libraries in a clean virtual environment, generate the file automatically:

```bash
pip freeze > requirements.txt
```

This captures every installed package, including indirect dependencies. It guarantees reproducibility but may include unnecessary bloat.

**For ML sprints, use Method 1.** You understand your stack and can specify it explicitly.

## Installing Dependencies from requirements.txt

Once the file exists, install all required packages using:

```bash
pip install -r requirements.txt
```

This command:
1. Reads each line of the requirements file
2. Installs the specified package version
3. Places everything in the active virtual environment

Always use a virtual environment first, then install from requirements.txt:

```bash
python -m venv venv
source venv/bin/activate      # macOS/Linux
pip install -r requirements.txt
```

## What Is requirements.txt?

`requirements.txt` is a plain text file that lists Python packages your project needs and their exact versions.

Example:

```text
pandas==2.1.0
numpy==1.24.3
scikit-learn==1.3.0
joblib==1.3.1
matplotlib==3.7.2
seaborn==0.12.2
```

Each line specifies:
- Package name
- Version operator (== means exact version)
- Version number

The `==` operator is called **version pinning** and ensures everyone installs the same library version.

## Why Dependency Management Matters in ML

Code is only one part of an ML system. The environment is equally important.

Two developers running the same script can get different results if their library versions differ. In ML workflows, version changes can silently affect:

- Preprocessing behavior
- Model training and evaluation
- Numerical stability
- Serialization formats

Example version conflicts:

- A scikit-learn model trained in version 1.1 may fail to load in version 1.3
- A pandas minor update can change categorical data handling
- Default parameters may shift between versions

Without dependency management, your project works today but may fail tomorrow.

## Common Mistakes in Dependency Management

### 1. Installing Globally Without Isolation
Installing packages globally instead of using a virtual environment creates version conflicts between projects.

### 2. Forgetting to Update requirements.txt
Installing a new package but not adding it to `requirements.txt` causes import errors for teammates.

### 3. Hardcoding Versions Only Locally
Documenting versions only on your machine without sharing them in `requirements.txt` breaks reproducibility.

### 4. Not Testing on a Fresh Environment
If your project runs only on your machine, dependency management may be broken.

**Solution:** Before submission, delete your virtual environment and recreate it from scratch using `requirements.txt`. If it works, your dependencies are correct.

## Dependency Management and Model Reproducibility

Reproducibility in ML depends on multiple layers:

- Same dataset
- Same train/test split
- Same `random_state`
- Same hyperparameters
- **Same dependency versions**

If any of these changes, your results may change.

Dependency management is one of the most overlooked but most important parts of this chain.

Without pinned dependencies, your evaluation metrics may drift slightly over time — even if your code has not changed.

## Testing Reproducibility Before Submission

Before submitting your project, test if it can run from a clean environment:

1. Delete your virtual environment
   ```bash
   rm -rf venv    # macOS/Linux
   rmdir /s venv  # Windows
   ```

2. Create a fresh environment
   ```bash
   python -m venv venv
   source venv/bin/activate  # macOS/Linux
   ```

3. Install from requirements.txt
   ```bash
   pip install -r requirements.txt
   ```

4. Run your full pipeline
   ```bash
   python main.py
   ```

If everything works, your dependency management is correct. If it does not, fix it before submission.
