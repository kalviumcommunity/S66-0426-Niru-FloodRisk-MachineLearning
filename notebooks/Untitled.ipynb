## Module responsibilities

- `data_preprocessing.py`: clean data, handle missing values, split train/test
- `feature_engineering.py`: transform raw features into model-ready features
- `train.py`: fit the model and save artifacts
- `evaluate.py`: calculate metrics and create reports
- `predict.py`: load saved artifacts and predict on new data
- `config.py`: store paths, random seeds, and hyperparameters

This separation makes the code easier to debug, extend, and test.

## Recommended project structure

A practical ML project usually separates data, code, and outputs like this:

```text
project_root/
├── data/
│   ├── raw/
│   └── processed/
├── models/
├── reports/
├── logs/
├── src/
│   ├── __init__.py
│   ├── config.py
│   ├── data_preprocessing.py
│   ├── feature_engineering.py
│   ├── train.py
│   ├── evaluate.py
│   └── predict.py
├── requirements.txt
├── README.md
└── main.py
```

Each file should answer one clear question and avoid overlapping responsibilities.

````python
# Good: functions define behavior, main() controls execution
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier


def load_data(filepath):
    """Load and return raw data from CSV."""
    return pd.read_csv(filepath)


def clean_data(df):
    """Remove missing values from a DataFrame."""
    return df.dropna()


def split_data(df, target_column, test_size=0.2, random_state=42):
    """Split a DataFrame into train and test sets."""
    X = df.drop(target_column, axis=1)
    y = df[target_column]
    return train_test_split(X, y, test_size=test_size, random_state=random_state)


def train_model(X_train, y_train, random_state=42):
    """Train and return a Random Forest model."""
    model = RandomForestClassifier(random_state=random_state)
    model.fit(X_train, y_train)
    return model


def main():
    """Run the end-to-end training workflow."""
    df = load_data('data.csv')
    df = clean_data(df)
    X_train, X_test, y_train, y_test = split_data(df, 'target')
    model = train_model(X_train, y_train)
    score = model.score(X_test, y_test)
    print(f'Test accuracy: {score:.3f}')


if __name__ == '__main__':
    main()
````

## Why structure matters

As machine learning projects grow, notebooks alone become fragile. A clear file layout helps you:

- reuse preprocessing logic consistently
- train and evaluate multiple models without copy-paste
- load saved models for prediction without retraining
- test each component independently
- avoid silent bugs and data leakage

Poor structure is often the difference between an experiment and a maintainable system.

## Managing imports correctly

Use explicit imports so dependencies are easy to understand:

```python
from src.data_preprocessing import clean_data, split_data
from src.feature_engineering import build_preprocessing_pipeline
from src.config import RANDOM_STATE, MODEL_PATH
```

Avoid wildcard imports and circular imports. If two modules need the same constant or helper, move it to `config.py` or a shared utility module.

## Configuration management

Hardcoding file paths, seeds, and hyperparameters creates inconsistency. Put shared values in one place:

```python
RANDOM_STATE = 42
TEST_SIZE = 0.20
MODEL_PATH = "models/random_forest_v1.pkl"
PIPELINE_PATH = "models/preprocessing_pipeline.pkl"
```

Import those values wherever they are needed so the whole project stays aligned.

## Training and prediction should stay separate

Training fits the model. Prediction loads already fitted artifacts and applies them.

- training uses `fit()` or `fit_transform()`
- prediction uses `transform()` and `predict()`
- prediction should not retrain the model
- prediction should not refit preprocessing steps

This separation makes data leakage much harder to introduce.
